```python
def listdir(path, list_name):
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
                listdir(file_path, list_name)
        else:
                list_name.append(file_path)
```

```python
all_list = []
for f_row in f_list:
    line = f_row.decode("utf8")
    result = line[line.rfind('\t')+1:-2]
    if result!=None and re.match("军事",line)!=None:
        all_list.append(result)
        print(result)
len(all_list)
```

```python
import codecs
for path in list_name:
    print(path)
    file = open(path, 'rb').read().decode("utf8")

    '''
    正则匹配出url和content
    '''
    patternURL = re.compile(r'<url>(.*?)</url>', re.S)
    patternCtt = re.compile(r'<content>(.*?)</content>', re.S)

    classes = patternURL.findall(file)
    contents = patternCtt.findall(file)

    '''
    # 把所有内容小于30字符的文本全部过滤掉
    '''
    for i in range(contents.__len__())[::-1]:
        if len(contents[i]) < threh:
            contents.pop(i)
            classes.pop(i)

    '''
    把URL进一步提取出来，只提取出一级url作为类别
    '''
    for i in range(classes.__len__()):
        patternClass = re.compile(r'http://(.*?)/',re.S)
        classi = patternClass.findall(classes[i])
        classes[i] = classi[0]

    '''
    按照URL作为类别保存到samples文件夹中
    '''
    for i in range(classes.__len__()):
        file = './samples/' + classes[i] + '.txt'
        f = open(file,'a+',encoding='utf-8')
        f.write(contents[i]+'\n')   #加\n换行显示
```

# 随机划分训练集

```python
f = open("./搜狗新闻整理/总整理/myNews.txt","r",encoding='utf-8')           #源文件
fw = open("./搜狗新闻整理/总整理/train.txt","w+",encoding='utf-8')     #待写文件
```

```python
import random
raw_list = f.readlines()
random.shuffle(raw_list)
len(raw_list)
```

```python
count_item = 35000 
for i in range(count_item):                           #随机抽取数目 n 35000为训练集数目
    fw.writelines(raw_list[i])

f.close()
fw.close()
```

```python
fw_train = open("./搜狗新闻整理/总整理/train.txt","r",encoding='utf-8')
fw_test = open("./搜狗新闻整理/总整理/test.txt","w+",encoding='utf-8')     #待写文件，测试集
quit_text = fw_train.readlines()
for item in raw_list:
    if item not in quit_text:
        fw_test.writelines(item)
        
fw_train.close()
fw_test.close()
```
